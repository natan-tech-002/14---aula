# Notebook – Operador Ternário em Python


## Introdução

O operador ternário é uma forma simplificada de escrever estruturas condicionais em Python.

### Sintaxe

```python
valor_se_verdadeiro if condição else valor_se_falso
```


````python

idade = 20

resultado = "Maior de idade" if idade >= 18 else "Menor de idade"

print(resultado)
````

# Exercício 1 – Número Par ou Ímpar

Faça um programa que receba um número e informe se ele é par ou ímpar utilizando operador ternário.



````python

numero = int(input("Digite um número: "))

print("Par" if numero % 2 == 0 else "Ímpar")
````

# Exercício 2 – Maior ou Menor de Idade

Peça a idade do usuário e informe:
- “Maior de idade”
- “Menor de idade”

Utilize operador ternário.

````python
idade = int(input("Digite sua idade: "))

print("Maior de idade" if idade >= 18 else "Menor de idade")
````

# Exercício 3 – Maior Número

Receba dois números e mostre qual deles é o maior utilizando operador ternário.

````python

n1 = float(input("Digite o primeiro número: "))
n2 = float(input("Digite o segundo número: "))

print("O maior número é:", n1 if n1 > n2 else n2)
````

# Exercício 4 – Situação do Aluno

Peça a nota de um aluno e informe:
- “Aprovado” se nota ≥ 6

# Exercício 5 – Verificação de Senha

# Exercício 7 – Número Positivo ou Negativo

# Exercício 8 – Desconto em Compras

Peça o valor de uma compra.

Se o valor for maior ou igual a R$100, aplique 10% de desconto.
Caso contrário, mantenha o valor original.

````python
valor = float(input("Digite o valor da compra: R$ "))

final = valor * 0.9 if valor >= 100 else valor

print("Valor final: R$", final)

````

# Exercício 10 – Média do Aluno

Receba duas notas.

Calcule a média e informe:
- “Aprovado” se média ≥ 6
- “Recuperação” se média entre 4 e 5.9
- “Reprovado” se média < 4

````python
n1 = float(input("Digite a primeira nota: "))
n2 = float(input("Digite a segunda nota: "))

media = (n1 + n2) / 2

print(
    "Aprovado" if media >= 6
    else "Recuperação" if media >= 4
    else "Reprovado"
)

````
